```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
import librosa
import soundfile as sf
import gdown
from IPython.display import Audio, display

# 1. Unduh file audio dari Google Drive
file_id = '1eA1DzHNSOcQlMAsVswc960v9twNUSDpu'
url = f'https://drive.google.com/uc?id={file_id}'
output_audio_path = 'Through_The_Fire_And_Flames.mp3'
gdown.download(url, output_audio_path, quiet=False)

# 2. Baca file audio
y, sr = librosa.load(output_audio_path, sr=None)

# 3. Desain filter FIR low-pass dengan berbagai jendela
filter_order = 101
cutoff_freq = 0.3  # Normalized cutoff frequency (0.3 * Nyquist frequency)


windows = ['boxcar', 'hamming', 'hann', 'blackman', 'kaiser', 'bartlett', 'blackmanharris', 'tukey']

kaiser_beta = 8.6  # Beta value for Kaiser window

# Fungsi untuk mendesain filter dan mendapatkan respons frekuensi
def design_filter(window, beta=None):
    if window == 'kaiser' and beta is not None:
        b = signal.firwin(filter_order, cutoff=cutoff_freq, window=(window, beta), pass_zero=True)
    else:
        b = signal.firwin(filter_order, cutoff=cutoff_freq, window=window, pass_zero=True)
    w, h = signal.freqz(b)
    return b, w, h

# Plot respons frekuensi untuk semua jendela
plt.figure(figsize=(12, 8))
for window in windows:
    beta = kaiser_beta if window == 'kaiser' else None
    b, w, h = design_filter(window, beta)
    plt.plot(w / np.pi, 20 * np.log10(np.abs(h)), label=window)

plt.title('Frequency Response of FIR Low-pass Filters with Different Windows')
plt.xlabel('Normalized Frequency (×π rad/sample)')
plt.ylabel('Magnitude (dB)')
plt.legend()
plt.grid()
plt.show()
```

# IIR


## Low Pass


```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
import librosa
import soundfile as sf
import gdown
from IPython.display import Audio, display

# 1. Unduh file audio dari Google Drive
file_id = '1eA1DzHNSOcQlMAsVswc960v9twNUSDpu'
url = f'https://drive.google.com/uc?id={file_id}'
output_audio_path = 'Through_The_Fire_And_Flames.mp3'
gdown.download(url, output_audio_path, quiet=False)

# 2. Baca file audio
y, sr = librosa.load(output_audio_path, sr=None)

# 3. Desain filter IIR low-pass dengan berbagai jenis
filter_order = 4
cutoff_freq = 0.3  # Normalized cutoff frequency (0.3 * Nyquist frequency)

# Fungsi untuk mendesain filter IIR dan mendapatkan respons frekuensi
def design_iir_filter(filter_type, order, cutoff):
    if filter_type == 'butter':
        b, a = signal.butter(order, cutoff, btype='low', analog=False)
    elif filter_type == 'cheby1':
        b, a = signal.cheby1(order, 1, cutoff, btype='low', analog=False)
    elif filter_type == 'cheby2':
        b, a = signal.cheby2(order, 20, cutoff, btype='low', analog=False)
    elif filter_type == 'ellip':
        b, a = signal.ellip(order, 1, 20, cutoff, btype='low', analog=False)
    elif filter_type == 'bessel':
        b, a = signal.bessel(order, cutoff, btype='low', analog=False)
    else:
        raise ValueError("Unknown filter type")
    w, h = signal.freqz(b, a)
    return b, a, w, h

filter_types = ['butter', 'cheby1', 'cheby2', 'ellip', 'bessel']

# Plot respons frekuensi untuk semua filter
plt.figure(figsize=(12, 8))
for filter_type in filter_types:
    b, a, w, h = design_iir_filter(filter_type, filter_order, cutoff_freq)
    plt.plot(w / np.pi, 20 * np.log10(np.abs(h)), label=filter_type)

plt.title('Frequency Response of IIR Low-pass Filters with Different Designs')
plt.xlabel('Normalized Frequency (×π rad/sample)')
plt.ylabel('Magnitude (dB)')
plt.legend()
plt.grid()
plt.show()
```
